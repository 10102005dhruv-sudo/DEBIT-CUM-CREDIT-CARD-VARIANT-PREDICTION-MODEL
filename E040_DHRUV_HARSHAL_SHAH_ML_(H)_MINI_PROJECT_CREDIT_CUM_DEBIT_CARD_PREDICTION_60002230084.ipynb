```python
# Load the dataframes.
import pandas as pd
credit_df = pd.read_csv('/content/drive/MyDrive/ML HONOURS/ML MINI PROJECT/CREDIT CARD DATASET.csv')
debit_df = pd.read_csv('/content/drive/MyDrive/ML HONOURS/ML MINI PROJECT/DEBIT CARD DATASET.csv')

# Display the first 5 rows of each dataframe.
display(credit_df.head())
display(debit_df.head())

```

```python
import pandas as pd
import numpy as np
import gradio as gr
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
target_col_credit = 'card_variant'
target_col_debit = 'card_variant'

# We’ll use all columns except the target as input features
X_credit = credit_df.drop(columns=[target_col_credit])
y_credit = credit_df[target_col_credit]

X_debit = debit_df.drop(columns=[target_col_debit])
y_debit = debit_df[target_col_debit]
```

```python
Xc_train, Xc_test, yc_train, yc_test = train_test_split(X_credit, y_credit, test_size=0.2, random_state=42)
Xd_train, Xd_test, yd_train, yd_test = train_test_split(X_debit, y_debit, test_size=0.2, random_state=42)
```

```python
num_features = X_credit.select_dtypes(include=np.number).columns.tolist()
cat_features = X_credit.select_dtypes(exclude=np.number).columns.tolist()

numeric_transformer = Pipeline(steps=[('scaler', StandardScaler())])
categorical_transformer = Pipeline(steps=[('encoder', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, num_features),
        ('cat', categorical_transformer, cat_features)
    ]
)
```

```python
credit_model = Pipeline(steps=[('preprocessor', preprocessor),
                               ('classifier', RandomForestClassifier(n_estimators=150, random_state=42))])

debit_model = Pipeline(steps=[('preprocessor', preprocessor),
                              ('classifier', RandomForestClassifier(n_estimators=150, random_state=42))])

print("🔁 Training Credit Card Model...")
credit_model.fit(Xc_train, yc_train)
print("✅ Credit Model trained successfully!")

print("🔁 Training Debit Card Model...")
debit_model.fit(Xd_train, yd_train)
print("✅ Debit Model trained successfully!")

# Optional: Evaluate quickly
y_pred_credit = credit_model.predict(Xc_test)
y_pred_debit = debit_model.predict(Xd_test)
print("\n📊 CREDIT CARD CLASSIFICATION REPORT:")
print(classification_report(yc_test, y_pred_credit))
print("\n📊 DEBIT CARD CLASSIFICATION REPORT:")
print(classification_report(yd_test, y_pred_debit))
```

```python
import gradio as gr
import pandas as pd
import numpy as np
from collections import OrderedDict

# --- helper: extract training columns and defaults from a fitted pipeline's preprocessor ---
def extract_cols_and_defaults_from_pipeline(pipe):
    """
    Returns (full_columns_list, numeric_cols, categorical_cols, defaults_dict)
    Uses pipe.named_steps['preprocessor'].transformers_ which stores the original column lists.
    For categorical columns it tries to get encoder.categories_ to pick a sensible default.
    """
    if 'preprocessor' not in pipe.named_steps:
        raise RuntimeError("Pipeline does not contain 'preprocessor' step.")
    pre = pipe.named_steps['preprocessor']
    full_cols = []
    numeric_cols = []
    categorical_cols = []
    defaults = {}

    # transformers_ has tuples like (name, transformer, columns)
    for name, transformer, columns in pre.transformers_:
        if columns == 'drop' or columns is None:
            continue
        # extend full column list in the training order
        for c in columns:
            full_cols.append(c)

        if name == 'num':
            numeric_cols += list(columns)
            for c in columns:
                defaults[c] = 0.0
        else:
            # categorical transformer — attempt to read categories_ from the encoder
            categorical_cols += list(columns)
            # transformer may be a Pipeline (e.g., named_steps['encoder']) or an encoder directly
            enc = None
            if hasattr(transformer, 'named_steps'):
                # pipeline -> try to find OneHotEncoder step
                for step_name, step in transformer.named_steps.items():
                    if hasattr(step, 'categories_'):
                        enc = step
                        break
            elif hasattr(transformer, 'categories_'):
                enc = transformer

            if enc is not None and hasattr(enc, 'categories_'):
                # enc.categories_ is a list of arrays in the same order as 'columns'
                # if lengths align, assign first category as default for each column
                cats = enc.categories_
                if len(cats) == len(columns):
                    for col_name, cat_arr in zip(columns, cats):
                        defaults[col_name] = (cat_arr[0] if len(cat_arr) > 0 else "Unknown")
                else:
                    # fallback: use the first available category for all cols (not ideal, but safe)
                    fallback = cats[0][0] if len(cats) and len(cats[0]) else "Unknown"
                    for col_name in columns:
                        defaults[col_name] = fallback
            else:
                # no encoder metadata — use 'Unknown'
                for c in columns:
                    defaults[c] = "Unknown"

    return full_cols, numeric_cols, categorical_cols, defaults

# --- Try to extract for both models (credit_model and debit_model must exist) ---
try:
    credit_full_cols, credit_num_cols, credit_cat_cols, credit_defaults = extract_cols_and_defaults_from_pipeline(credit_model)
except Exception as e:
    raise RuntimeError("Couldn't extract columns from credit_model. Make sure credit_model exists and has a trained preprocessor. Error: " + str(e))

try:
    debit_full_cols, debit_num_cols, debit_cat_cols, debit_defaults = extract_cols_and_defaults_from_pipeline(debit_model)
except Exception as e:
    # If debit_model not available, set to None (we'll only support credit)
    debit_full_cols = None
    debit_num_cols = []
    debit_cat_cols = []
    debit_defaults = {}
    print("Warning: couldn't extract from debit_model — debit predictions will be disabled. Error:", e)

# Merge defaults strategy: prefer credit_defaults (or you can choose to use debit defaults for debit model)
# We'll use credit_defaults for autofill whenever we need defaults (common training schema assumed similar).
global_full_cols = credit_full_cols  # columns order used for prediction inputs for both models

# --- safe input -> full-row builder using defaults ---
def build_full_input_row(user_inputs: dict, full_cols: list, defaults: dict):
    """
    user_inputs: dict of the small set of user-provided keys (essential features).
    full_cols: list of column names expected by the pipeline (in order).
    defaults: dict with default values for columns (from encoder metadata when available).
    Returns: DataFrame with one row and columns in full_cols order, safe dtypes and no NaNs.
    """
    row = OrderedDict()
    for c in full_cols:
        if c in user_inputs and user_inputs[c] is not None:
            val = user_inputs[c]
            # coerce types for numeric-looking defaults
            if isinstance(defaults.get(c, None), float) or _looks_like_numeric(defaults.get(c, None)):
                # try to cast numeric
                try:
                    row[c] = float(val)
                except:
                    # fallback to mean-like numeric: 0.0
                    try:
                        row[c] = float(defaults.get(c, 0.0))
                    except:
                        row[c] = 0.0
            else:
                row[c] = str(val)
        else:
            # not provided by user -> use default if available otherwise safe fallback
            if c in defaults:
                row[c] = defaults[c]
            else:
                # choose fallback by name: numeric-like columns -> 0.0, else "Unknown"
                row[c] = 0.0 if c.lower().find('num')>=0 or c.lower().find('age')>=0 or c.lower().find('income')>=0 or c.lower().find('score')>=0 else "Unknown"
    # final convert to DataFrame
    df = pd.DataFrame([row], columns=full_cols)
    # replace nulls
    df = df.fillna(0)
    return df

# helper to check numeric-ish defaults
def _looks_like_numeric(x):
    try:
        float(x)
        return True
    except:
        return False

# --- Prediction function using small user input (only essentials) ---
def predict_card_type_after_training(card_choice, age, monthly_income, cibil_score, education, occupation,
                                     marital_status, num_dependents, spending_score, num_credit_cards, city_tier):
    """
    This function requires credit_model and debit_model already trained and in memory.
    It accepts only essential inputs, builds full row, and predicts.
    """
    # safe-cast and normalize input keys to match training column names if needed
    # Map common alias names user sees -> actual column names in training set
    # We'll attempt to match case-insensitively
    user_small = {
        # try to map names - adapt keys if your training columns differ (e.g., 'monthly_income' vs 'monthly_income')
        'age': age,
        'monthly_income': monthly_income,
        'cibil': cibil_score if 'cibil' in global_full_cols else cibil_score,  # keep both attempts
        'cibil_score': cibil_score,
        'education': education,
        'occupation': occupation,
        'marital_status': marital_status,
        'num_dependents': num_dependents,
        'spending_score': spending_score,
        'num_credit_cards': num_credit_cards,
        'city_tier': city_tier
    }

    # Build full rows for credit & debit using defaults
    credit_input_df = build_full_input_row(user_small, credit_full_cols, credit_defaults)
    if debit_full_cols is not None:
        debit_input_df = build_full_input_row(user_small, debit_full_cols, debit_defaults)
    else:
        debit_input_df = None

    out_lines = []
    # Predict credit
    if card_choice in ["Credit Card", "Both"]:
        try:
            pred_credit = credit_model.predict(credit_input_df)[0]
            # try to get probabilities if available
            probs = None
            if hasattr(credit_model, 'predict_proba'):
                try:
                    probs = credit_model.predict_proba(credit_input_df)[0]
                    classes = credit_model.named_steps['clf'].classes_ if 'clf' in credit_model.named_steps else credit_model.named_steps['classifier'].classes_
                    top3 = sorted(zip(classes, probs), key=lambda x: x[1], reverse=True)[:3]
                except Exception:
                    top3 = None
            else:
                top3 = None
            out_lines.append("**Credit Card Recommendation**")
            out_lines.append(f"Top pick: {pred_credit}")
            if top3:
                out_lines.append("Top 3 with prob:")
                for nm,p in top3:
                    out_lines.append(f"- {nm}: {p*100:.1f}%")
        except Exception as e:
            out_lines.append("Credit model error: " + str(e))

    # Predict debit
    if card_choice in ["Debit Card", "Both"] and debit_input_df is not None:
        try:
            pred_debit = debit_model.predict(debit_input_df)[0]
            probs = None
            if hasattr(debit_model, 'predict_proba'):
                try:
                    probs = debit_model.predict_proba(debit_input_df)[0]
                    classes = debit_model.named_steps['clf'].classes_ if 'clf' in debit_model.named_steps else debit_model.named_steps['classifier'].classes_
                    top3 = sorted(zip(classes, probs), key=lambda x: x[1], reverse=True)[:3]
                except Exception:
                    top3 = None
            else:
                top3 = None
            out_lines.append("\n**Debit Card Recommendation**")
            out_lines.append(f"Top pick: {pred_debit}")
            if top3:
                out_lines.append("Top 3 with prob:")
                for nm,p in top3:
                    out_lines.append(f"- {nm}: {p*100:.1f}%")
        except Exception as e:
            out_lines.append("Debit model error: " + str(e))

    return "\n".join(out_lines)

# --- GRADIO UI (launch) ---
iface = gr.Interface(
    fn=predict_card_type_after_training,
    inputs=[
        gr.Radio(["Credit Card","Debit Card","Both"], label="Predict for"),
        gr.Number(label="Age", value=30),
        gr.Number(label="Monthly Income (₹)", value=60000),
        gr.Number(label="CIBIL Score (300–900)", value=720),
        gr.Dropdown(["NoFormal","HighSchool","Graduate","Postgraduate","Doctorate"], label="Education", value="Graduate"),
        gr.Dropdown(["Salaried","Self-employed","Student","Retired","Unemployed","Housewife"], label="Occupation", value="Salaried"),
        gr.Dropdown(["Single","Married","Divorced"], label="Marital Status", value="Single"),
        gr.Number(label="Number of Dependents", value=0),
        gr.Slider(1,100, step=1, label="Spending Score", value=50),
        gr.Number(label="Number of Existing Credit Cards", value=1),
        gr.Dropdown(["Tier1","Tier2","Tier3"], label="City Tier", value="Tier1")
    ],
    outputs=gr.Markdown(),
    title="Card Variant Predictor (post-training)",
    description="App uses trained pipelines credit_model and debit_model already in memory. Enter essential fields and click Submit."
)

iface.launch(share=True, debug=True)
```
